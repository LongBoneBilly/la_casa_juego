```python
import random

class Character:
    def __init__(self):
        self.HP = 100
        self.Strength = 80
        self.Rage = 50
        self.Knowledge = 100
        self.Speed = 75
        self.Experience = 60
        self.Attack = 90

    def display_stats(self):
        print(f"HP: {self.HP}, Strength: {self.Strength}, Rage: {self.Rage}, Knowledge: {self.Knowledge}, Speed: {self.Speed}, Experience: {self.Experience}, Attack: {self.Attack}")

class Mafins(Character):
    def __init__(self):
        super().__init__()
        self.name = "Mafins"

    def choose_attributes(self):
        print(f"Welcome {self.name}! Choose additional attributes:")
        self.Attack += 10
        self.Experience += 15
        self.Rage += 20

class Jefe(Character):
    def __init__(self):
        super().__init__()
        self.name = "Jefe"
    def choose_attributes(self):
        print(f"Welcome {self.name}! Choose additional attributes:")
        self.Attack += 10
        self.Experience += 15
        self.Rage += 20

class OtherCharacters(Character):
    def __init__(self, name):
        super().__init__()
        self.name = name

    def choose_attributes(self):
        print(f"Welcome {self.name}! Choose additional attributes:")
        self.Speed += 20
        self.Knowledge += 15
        self.HP += 10
        self.Strength += 15

class Enemy:
    def __init__(self, name, HP, Attack):
        self.name = name
        self.HP = HP
        self.Attack = Attack

    def display_stats(self):
        print(f"{self.name} - HP: {self.HP}, Attack: {self.Attack}")

def choose_weapon():
    weapons = ["frying pan", "wet towel", "chancla", "leather belt", "coat hanger", "water cannon", "gardening hoe", "pencils", "hammer", "jigsaw", "fists", "keyboard", "raspberry pi kit"]
    print("Choose your weapon:")
    for i, weapon in enumerate(weapons, start=1):
        print(f"{i}. {weapon}")

    choice = input("Enter the number of your choice, or enter 'random' for a random weapon: ")
    if choice.lower() == "random":
        return random.choice(weapons)
    elif choice.isdigit() and 1 <= int(choice) <= len(weapons):
        return weapons[int(choice) - 1]
    else:
        print("Invalid choice. Picking a random weapon.")
        return random.choice(weapons)

def use_health_potion(player):
    if player.HP < 50:
        print("You have used a health potion.")
        player.HP += 30
        if player.HP > 100:
            player.HP = 100
        print(f"Your HP is now {player.HP}.")
    else:
        print("You don't need a health potion right now.")

def main():
    print("Choose your character:")
    print("1. Mafins")
    print("2. Jefe")
    print("3. Manchis")
    print("4. Tipiti")
    print("5. Panzon")
    print("6. Lagartijo")

    choice = input("Enter the number of your choice: ")
    if choice == "1":
        player = Mafins()
    elif choice == "2":
        player = Jefe()
    elif choice == "3" or choice == "4" or choice == "5" or choice == "6":
        player = OtherCharacters(["Manchis", "Tipiti", "Panzon", "Lagartijo"][int(choice) - 3])
    else:
        print("Invalid choice. Exiting...")
        return

    player.choose_attributes()
    weapon = choose_weapon()

    print(f"Congratulations! You are now playing as {player.name} with the following stats:")
    player.display_stats()
    print(f"You have chosen a {weapon} as your weapon.")

    for round_num in range(1, 4):
        lolo = Enemy("Lolo the Man-eating Rabbit", 300, 100)
        print(f"\nRound {round_num}: Prepare yourself! You have encountered {lolo.name}.")
        lolo.display_stats()

        input("Press Enter to start the battle...")

        # Player gets +50 attack boost against Lolo
        player.Attack += 50

        while player.HP > 0 and lolo.HP > 0:
            print("1. Attack")
            print("2. Use Health Potion")
            choice = input("Enter the number of your choice: ")

            if choice == "1":
                # Player's turn
                player_attack = player.Attack + random.randint(-10, 10)
                lolo.HP -= player_attack
                print(f"You dealt {player_attack} damage to {lolo.name}.")

                if lolo.HP <= 0:
                    print(f"Congratulations! You defeated {lolo.name} in Round {round_num}.")
                    break

                # Lolo's turn
                lolo_attack = lolo.Attack + random.randint(-5, 5)
                player.HP -= lolo_attack
                print(f"{lolo.name} dealt {lolo_attack} damage to you.")
            elif choice == "2":
                use_health_potion(player)
            else:
                print("Invalid choice. Try again.")

            if player.HP <= 0:
                print(f"You were defeated by {lolo.name}. Game over.")
                return

            print(f"Your HP: {player.HP}, {lolo.name}'s HP: {lolo.HP}")

if __name__ == "__main__":
    main()
```
